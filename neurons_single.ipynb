#### Input
Get the x and y position of the player and the fruit

```python
import math
import pandas as pd
import numpy as np

raw_values = [2, 4, 6, 8] # (x_p: 2, y_p: 4), (x_f: 6, y_f: 8)
```

Generate the initial weighs and bias for each layer of neurons.
These are the "matrixes" all AIs run on.

```python
def generate_random_weights(rows, cols):
    return np.random.uniform(-1, 1, (rows, cols))


def generate_random_biases(size):
    return np.random.uniform(-1, 1, size)


layer_1_weights = generate_random_weights(3, 4)
layer_1_bias = generate_random_biases(3)
layer_2_weights = generate_random_weights(2, 3)
layer_2_bias = generate_random_biases(2)

print("z_in w + b:"); print(pd.DataFrame(np.column_stack([layer_1_weights, layer_1_bias]), columns=["w1", "w2", "w3", "w4","bias"], index=["n1", "n2", "n3"]))
print("\nz_out w + b:"); print(pd.DataFrame(np.column_stack([layer_2_weights, layer_2_bias]), columns=["w1", "w2", "w3","bias"], index=["out1", "out2"]))
```

#### 1st hidden layer
Theres 3 neurons on this hidden layer. 
For each neuron, we multiply all inputs by 4 weights and a bias:  
$$Z1 = (x1 * w1) + (x2 * w2) + (x3 * w3) + (x4 * w4) + b1$$
After that we need to normlize the neuron's value to be betwen -1 and 1. For that we use tanh:
$$ tanh(Zn) <=> \frac{(e^z - e^{-z})}{(e^z + e^{-z})} $$
The output of the neuron will be wathever the tanh() is.

```python
Z_layer_1 = []
for neuron_weights in layer_1_weights:  # for each row of weights
    z_in_row = sum(n * w for n, w in zip(raw_values, neuron_weights)) # multiply and sum the values and weights
    Z_layer_1.append(z_in_row)

Z_layer_1 = [v + b for v, b in zip(Z_layer_1, layer_1_bias)] # add bias

print("Z_in:"); print(pd.DataFrame(Z_layer_1, columns=["n_Z_in"], index=["n1", "n2", "n3"]))

A_layer_1 = [math.tanh(v) for v in Z_layer_1] # normalize

print("\nA_in:"); print(pd.DataFrame(A_layer_1, columns=["n_A_in"], index=["n1", "n2", "n3"]))
```

#### 2nd hidden layer
Same thing, only the number of neurons and inputs change, to 2x3

```python
Z_layer_2 = []
for neuron_weights in layer_2_weights:  # for each row of weights
    Z_out_row = sum(n * w for n, w in zip(A_layer_1, neuron_weights)) # multiply and sum the values and weights
    Z_layer_2.append(Z_out_row)

Z_layer_2 = [v + b for v, b in zip(Z_layer_2, layer_2_bias)] # add bias

print("Z_out:"); print(pd.DataFrame(Z_layer_2, columns=["n_Z_out"], index=["out1", "out2"]))

A_layer_2 = [math.tanh(v) for v in Z_layer_2] # normalize

print("\nA_out:"); print(pd.DataFrame(A_layer_2, columns=["n_A_out"], index=["out1", "out2"]))
```

#### Output
Will return the angle the player must turn and how fast to move in that direction

```python
angle = (A_layer_2[0] + 1) / 2 * 360
speed = (A_layer_2[1] + 1) / 2 * 5

print("Angle:", angle)
print("Speed:", speed)
```
